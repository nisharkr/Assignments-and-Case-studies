```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading the cars data csv

```python
cars_data=pd.read_csv('cars_data.csv')
cars_data.head()
```

# Checking for null values in the data

```python
cars_data.info()  #shows 10000 entries non null
```

# Bar graph : Male Vs Female buyers

```python
#x=cars_data.groupby('Buyer Gender').count()
#x=cars_data['Buyer Gender'].value_counts()
#x["Male"]
cars_data['Buyer Gender'].value_counts().plot(kind='bar',figsize=(7,6),rot=0);
plt.xlabel("Gender",labelpad=10)
plt.ylabel("Count of Buyers")
plt.title("Male vs Female Buyers")
```

# Top 5 cars based on sales price

```python
dfTop5Cars = cars_data.sort_values(by='Sale Price',ascending=False).head(5)#[["Make","Model","Sale Price"]]
dfTop5Cars = dfTop5Cars.reset_index(drop=True)
dfTop5Cars.index=np.arange(1,len(dfTop5Cars)+1)
dfTop5Cars
```

# Least 5 cars based on Resell price

```python
dfLeast5Cars = cars_data.sort_values(by='Resell Price').head(5)
dfLeast5Cars = dfLeast5Cars.reset_index(drop=True)
dfLeast5Cars.index=np.arange(1,len(dfLeast5Cars)+1)
dfLeast5Cars
```
